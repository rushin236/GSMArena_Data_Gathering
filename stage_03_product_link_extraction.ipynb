```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv(
    "./data/gsmarena_product_pages.csv",
)
df
```

```python
data = []
```

```python
def get_prod_links(brand, devices, link, code, gen_link, html):
    soup = BeautifulSoup(html, "html.parser")
    tags = soup.select("div.makers ul li")
    if len(tags) != 0:
        for tag in tags:
            link = tag.select_one("a").get("href")
            prefix = "https://www.gsmarena.com/"
            prod_link = prefix + link
            img_link = tag.select_one("img").get("src")
            name = tag.select_one("span").get_text()
            data.append(
                (brand, name, prod_link, img_link, devices, link, code, gen_link, html)
            )
            print(f"name: {name} link:{prod_link} img_link:{img_link}")
```

```python
for brand, devices, link, code, gen_link, html in df.values:
    get_prod_links(brand, devices, link, code, gen_link, html)
```

```python
index = []
for idx, html in enumerate(df["html"]):
    soup = BeautifulSoup(html, "lxml")
    tags = soup.select("div.makers ul li")
    if len(tags) == 0:
        print(idx)
        index.append(idx)
```

```python
data_df = pd.DataFrame(
    data,
    columns=[
        "brand",
        "name",
        "prod_link",
        "img_link",
        "devices",
        "link",
        "code",
        "gen_link",
        "html",
    ],
)
data_df.head()
```

```python
data_df.to_csv(index=False, path_or_buf="./data/gsmarena_product_link.csv")
```
